## Client Application to Consume web service

```python
!pip install widgetsnbextension
!pip install ipywidgets
import urllib.request as req
import requests
from ipywidgets import Label, BoundedFloatText, BoundedIntText, Dropdown, Button, Output, VBox, interact, interactive

predict_label = Label('Prediction of an Individual with account ownership in a Developing country')
year_text = BoundedIntText(min=2016, max=2018, value=2017, description="Year:", disabled=False)
location_type_dropdown = Dropdown(options=['Rural', 'Urban'], description='Location Type:', disabled=False)
cellphone_access_dropdown = Dropdown(options=['Yes', 'No'], value="Yes", description='Cellphone Access:', disabled=False)
household_size_text = BoundedIntText(min=1, max=21, value=3, description="Household Size:", disabled=False)
age_text = BoundedIntText(min=16, max=100, value=24, description="Age:", disabled=False)
relationship_with_head_dropdown = Dropdown(options=['Spouse', 'Head of Household', 'Other relative', 'Child', 'Parent',
                                                    'Other non-relatives'], description='Relationship with Head:',
                                           disabled=False)

marital_status_dropdown = Dropdown(options=['Married/Living together', 'Widowed', 'Single/Never Married',
                                            'Divorced/Seperated', 'Dont know'], description='Marital Status:',
                                   disabled=False)

education_level_dropdown = Dropdown(options=['Other/Dont know/RTA', 'No formal education', 'Primary education',
                                             'Vocational/Specialised training', 'Secondary education',
                                             'Tertiary education'], description='Education Status (Highest):',
                                    disabled=False)

job_type_dropdown = Dropdown(options=['Self employed', 'Government Dependent', 'Formally employed Private',
                                      'Informally employed', 'Formally employed Government', 'Farming and Fishing',
                                      'Remittance Dependent', 'Other Income', 'Dont Know/Refuse to answer',
                                      'No Income'], description='Job Type:', disabled=False)
gender_dropdown = Dropdown(options=['Female', 'Male'], description='Gender:', disabled=False)
country_dropdown = Dropdown(options=['Kenya', 'Rwanda', 'Tanzania', 'Uganda'], description='Country:', disabled=False)

predict_button = Button(description="Predict")
predict_output = Output()


# Button click event handlers ...
def predict_button_on_click(b):

    relationship_with_head_dropdown.value = f"Other non-relatives"
    param6 = req.pathname2url(relationship_with_head_dropdown.value)
    
    marital_status_dropdown.value = 'Married/Living together'
    params7 = req.pathname2url(marital_status_dropdown.value)

    education_level_dropdown.value = 'Other/Dont know/RTA'
    params8 = req.pathname2url(education_level_dropdown.value)

    job_type_dropdown.value = 'Dont Know/Refuse to answer'
    params9 = req.pathname2url(job_type_dropdown.value)


    endpoint = f"https://bank-class-app.azurewebsites.net/bank?"

    request_url = endpoint + f'Year={year_text.value}' + f'&Location_type={location_type_dropdown.value}' + f'&Cellphone_access={cellphone_access_dropdown.value}' + f'&Household_size={household_size_text.value}' + f'&Age={age_text.value}' + '&Relationship_with_head=' + param6 + '&Marital_status=' + params7 + '&Education_level=' + params8 + '&Job_type=' + params9 + f'&Gender_female={gender_dropdown.value}' + f'&Country={country_dropdown.value}'
    request_url

    response = requests.get(request_url)
    account_exists = response.json()["account_exists"]

    predict_output.clear_output()
    with predict_output:
        print(f"The individual is predicted to have an account?: {account_exists}")


predict_button.on_click(predict_button_on_click)

vbox_predict = VBox(
    [predict_label, year_text, location_type_dropdown, cellphone_access_dropdown, household_size_text, age_text,
     relationship_with_head_dropdown, marital_status_dropdown, education_level_dropdown, job_type_dropdown,
     gender_dropdown, country_dropdown, predict_button, predict_output])

vbox_predict
```
